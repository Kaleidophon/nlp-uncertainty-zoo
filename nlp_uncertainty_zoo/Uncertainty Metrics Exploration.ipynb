```python
### Test AUPR and AUROC
y_true = [0] * 10 + [1] * 10
y_pred = np.concatenate([np.random.randn(10) * 2 - 2, np.random.randn(10) * 2 + 2])
print(metrics.aupr(y_true, y_pred))
print(metrics.auroc(y_true, y_pred))
```

```python
### Test Kendall's tau
y_losses = y_pred - 5
print(metrics.kendalls_tau(y_losses, y_pred))
print(metrics.kendalls_tau(np.random.rand(20) - 10, y_pred))
```
